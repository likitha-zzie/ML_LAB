```python
import pandas as pd
df=pd.read_csv("/content/Dataset of Diabetes .csv")
df2=pd.read_csv("/content/adult.csv")
df.info()
df2.info()
print(df.describe())
print(df2.describe())
```

```python
import pandas as pd
import numpy as np

# ---------------------------
# Handling Missing Values
# ---------------------------

# Check for missing values in df
missing_df = df.isnull().sum()
print("Missing values in df:\n", missing_df)

# For df2
missing_df2 = df2.isnull().sum()
print("Missing values in df2:\n", missing_df2)

# Impute or drop missing values
# Option 1: Drop rows with missing values (if the number of missing values is small)
df = df.dropna()
df2 = df2.dropna()

# Option 2: Impute missing values (example: filling with the median for numerical columns)

numeric_columns_df = df.select_dtypes(include=np.number).columns
numeric_columns_df2 = df2.select_dtypes(include=np.number).columns

df[numeric_columns_df] = df[numeric_columns_df].fillna(df[numeric_columns_df].median())
df2[numeric_columns_df2] = df2[numeric_columns_df2].fillna(df2[numeric_columns_df2].median())

# ---------------------------
# Handling Categorical Data
# ---------------------------

cat_columns_df = df.select_dtypes(include=['object']).columns
cat_columns_df2 = df2.select_dtypes(include=['object']).columns

df = pd.get_dummies(df, columns=cat_columns_df, drop_first=True)
df2 = pd.get_dummies(df2, columns=cat_columns_df2, drop_first=True)

# ---------------------------
# Handling Outliers
# ---------------------------

# For df, calculate IQR for numeric columns only
Q1_df = df[numeric_columns_df].quantile(0.25)
Q3_df = df[numeric_columns_df].quantile(0.75)
IQR_df = Q3_df - Q1_df

# Identify outliers in df
outliers_df = ((df[numeric_columns_df] < (Q1_df - 1.5 * IQR_df)) | (df[numeric_columns_df] > (Q3_df + 1.5 * IQR_df))).sum()
print("Outliers in df:\n", outliers_df)

# Removing outliers by dropping rows that contain outliers
df = df[~((df[numeric_columns_df] < (Q1_df - 1.5 * IQR_df)) | (df[numeric_columns_df] > (Q3_df + 1.5 * IQR_df))).any(axis=1)]

# For df2, calculate IQR for numeric columns only
numeric_columns_df2 = df2.select_dtypes(include=np.number).columns  # Ensure we only use numeric columns
Q1_df2 = df2[numeric_columns_df2].quantile(0.25)
Q3_df2 = df2[numeric_columns_df2].quantile(0.75)
IQR_df2 = Q3_df2 - Q1_df2

# Identify outliers in df2
outliers_df2 = ((df2[numeric_columns_df2] < (Q1_df2 - 1.5 * IQR_df2)) | (df2[numeric_columns_df2] > (Q3_df2 + 1.5 * IQR_df2))).sum()
print("Outliers in df2:\n", outliers_df2)

# Removing outliers by dropping rows that contain outliers
df2 = df2[~((df2[numeric_columns_df2] < (Q1_df2 - 1.5 * IQR_df2)) | (df2[numeric_columns_df2] > (Q3_df2 + 1.5 * IQR_df2))).any(axis=1)]

# Final info after cleaning
df.info()
df2.info()

# Final statistics after cleaning
print(df.describe())
print(df2.describe())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Assuming you have a dataframe `df` and `df2`

# -------------------------------
# Min-Max Scaling (Normalization)
# -------------------------------

# Initialize MinMaxScaler
min_max_scaler = MinMaxScaler()

# Apply Min-Max Scaling to the numeric columns in df
df[numeric_columns_df] = min_max_scaler.fit_transform(df[numeric_columns_df])

# Apply Min-Max Scaling to the numeric columns in df2
df2[numeric_columns_df2] = min_max_scaler.fit_transform(df2[numeric_columns_df2])

# Check the transformed df and df2
print("Transformed df after Min-Max Scaling:\n", df.head())
print("Transformed df2 after Min-Max Scaling:\n", df2.head())

# -------------------------------
# Standard Scaling
# -------------------------------

# Initialize StandardScaler
standard_scaler = StandardScaler()

# Apply Standard Scaling to the numeric columns in df
df[numeric_columns_df] = standard_scaler.fit_transform(df[numeric_columns_df])

# Apply Standard Scaling to the numeric columns in df2
df2[numeric_columns_df2] = standard_scaler.fit_transform(df2[numeric_columns_df2])

# Check the transformed df and df2
print("Transformed df after Standard Scaling:\n", df.head())
print("Transformed df2 after Standard Scaling:\n", df2.head())
```
